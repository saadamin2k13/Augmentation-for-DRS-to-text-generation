**generating a case-based dictionary for DRS name changes.**

```python
# Open the input file in read mode
with open('dict.txt', 'r') as f:
    # Read the content of the file and split it into lines
    lines = f.read().splitlines()

# Initialize an empty list to store the output lines
output_lines = []

# Loop through the input lines and generate the output lines
for line in lines:
    # Split the line into key and value
    key, value = line.split(':')

    # Generate the output lines for the original key-value pair and the lowercase key-value pair
    output_lines.append(f"{key}:{value.strip()} {key.lower()}:{value.lower().strip()} \n")

# Open the output file in write mode and write the output lines to it
with open('case_dict.txt', 'w') as f:
    f.write(''.join(output_lines))
```
